## Resize Images ##

The images were quite large, so I needed to resize them in order to be able to process 


```python
import os
import cv2

# Set the input folder path
folder_path = 'frames/'

# Set the output folder path
output_folder_path = 'resized_images3/'

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Get the list of image files in the input folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Loop through each image file and resize
for file_name in image_files:
    try:
        # Read the image
        image_path = os.path.join(folder_path, file_name)
        image = cv2.imread(image_path)

        if image is not None:
            # Set the desired width and height
            desired_width = 200
            desired_height = 100

            # Resize the image
            resized_image = cv2.resize(image, (desired_width, desired_height))

            # Write the resized image to the output folder
            output_path = os.path.join(output_folder_path, file_name)
            cv2.imwrite(output_path, resized_image)

            print(f"Resized image '{file_name}' and saved to '{output_path}'")
        else:
            print(f"Skipping image '{file_name}' due to loading error")
    except Exception as e:
        print(f"An error occurred while processing image '{file_name}': {str(e)}")
```

## Train the VAE Model ##

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras import backend as K

# Assuming image_data is already loaded and preprocessed
# Normalize the image data
image_data = image_data / 255.0

# Define the dimensions of the images
image_height, image_width, num_channels = image_data.shape[1:]

# Define the latent space dimension (the size of the bottleneck layer)
latent_dim = 10

# Define the encoder
encoder_inputs = keras.Input(shape=(image_height, image_width, num_channels))
x = layers.Flatten()(encoder_inputs)
x = layers.Dense(100, activation="relu")(x)
encoder_outputs = layers.Dense(latent_dim)(x)
encoder = keras.Model(encoder_inputs, encoder_outputs, name="encoder")

# Define the decoder
decoder_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(100, activation="relu")(decoder_inputs)
x = layers.Dense(image_height * image_width * num_channels, activation="sigmoid")(x)
decoder_outputs = layers.Reshape((image_height, image_width, num_channels))(x)
decoder = keras.Model(decoder_inputs, decoder_outputs, name="decoder")

tf.keras.backend.clear_session()

# Define the VAE as a combined model
vae_inputs = encoder_inputs
vae_outputs = decoder(encoder_outputs)
vae = keras.Model(vae_inputs, vae_outputs, name="vae")

def vae_loss(inputs, outputs):
    reconstruction_loss = keras.losses.binary_crossentropy(inputs, outputs)
    reconstruction_loss *= image_height * image_width * num_channels
    kl_loss = 1 + K.log(K.square(K.exp(outputs))) - K.square(outputs)
    kl_loss = K.sum(kl_loss, axis=-1)
    kl_loss *= -0.5
    vae_loss = K.mean(reconstruction_loss + kl_loss)
    return vae_loss

vae.compile(optimizer=keras.optimizers.Adam(), loss=vae_loss)
vae.fit(image_data, image_data, epochs=10, batch_size=20)

# Generate a new image
random_latent_vector = np.random.normal(size=(1, latent_dim))
generated_image = decoder.predict(random_latent_vector)

# Rescale the generated image to the original range
generated_image = generated_image * 255.0
generated_image = generated_image.astype(np.uint8)
```

## Time to make GIFs ##

Because my final output was going to be a video, I decided to first create gifs using images created through the random latent vectors. These are some of my favorite things produced from this project.

```python
import numpy as np
import imageio

# Define the number of GIFs and frames
num_gifs = 6
num_frames = 25

# Generate random latent vectors for each GIF
latent_vectors = np.random.normal(size=(num_gifs, num_frames, latent_dim))

# Create a folder to save the frames
output_folder = "gif_frames"
os.makedirs(output_folder, exist_ok=True)

# Generate frames for each GIF
for gif_idx in range(num_gifs):
    gif_frames = []
    for frame_idx in range(num_frames):
        # Get the latent vector for the current frame
        latent_vector = latent_vectors[gif_idx, frame_idx]

        # Generate an image from the latent vector using the decoder
        generated_image = decoder.predict(np.array([latent_vector]))[0]

        # Rescale the generated image to the original range
        generated_image = generated_image * 255.0
        generated_image = generated_image.astype(np.uint8)

        # Append the generated image to the frames list
        gif_frames.append(generated_image)

    # Animate the latent vector by interpolating between frames
    interpolated_frames = []
    for i in range(len(gif_frames) - 1):
        for j in range(10):
            interpolated_frame = (gif_frames[i] * (10 - j) + gif_frames[i + 1] * j) / 10
            interpolated_frames.append(interpolated_frame)

    # Save the frames as a GIF
    output_file = os.path.join(output_folder, f"gif_{gif_idx}.gif")
    imageio.mimsave(output_file, interpolated_frames, fps=10)

print("GIFs generated successfully!")
```


## Now to the music... ##

For this I imported a song I'd written (simply because it was saved to my computer). Then I experimented with the python librosa library. I decided that it would be interesting to see if anything in the video changes based on the amplitude (loudness) of the song.

Because I knew I wanted to use the amplitude values in the code to create a video, I downloaded them all in a csv on my computer.

```python
import numpy as np
import pandas as pd
import librosa

# Load the audio file
audio_path = 'syd.m4a'
audio, sr = librosa.load(audio_path)

# Calculate the RMS manually
rms = np.sqrt(np.mean(np.square(audio)))

# Convert RMS to dB using NumPy
loudness = 20 * np.log10(rms)

# Define the desired sampling rate for loudness measurements (30 times per second)
loudness_sampling_rate = 30

# Calculate the number of frames needed
num_frames = int(len(audio) / sr * loudness_sampling_rate)

# Initialize an array to store the loudness values
loudness_values = np.zeros(num_frames)

# Calculate loudness at each frame
for frame_idx in range(num_frames):
    start_sample = int(frame_idx * sr / loudness_sampling_rate)
    end_sample = int((frame_idx + 1) * sr / loudness_sampling_rate)
    frame_audio = audio[start_sample:end_sample]
    frame_rms = np.sqrt(np.mean(np.square(frame_audio)))
    loudness_values[frame_idx] = 20 * np.log10(frame_rms)

# Create a time array for the loudness values
times = np.arange(num_frames) / loudness_sampling_rate

# Create a pandas DataFrame to store the loudness values
data = pd.DataFrame({'Time': times, 'Loudness': loudness_values})

# Save the data to a CSV file
data.to_csv('loudness_data.csv', index=False)

# Display the loudness values in a grid format
print(data)
```
